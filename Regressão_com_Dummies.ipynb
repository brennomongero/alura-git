```python
import pandas as pd
import pyarrow.feather as feather
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
corrupcao = pd.read_feather('corrupcao.feather')
corrupcao
```

```python
corrupcao.describe(include='all')
```

```python
corrupcao.info()
```

```python
corrupcao.regiao.value_counts()
```

```python
#Criando Dummies
corrupcao_dummies = pd.concat([corrupcao, pd.get_dummies(corrupcao['regiao'],)], axis=1).drop(columns=['regiao'])
corrupcao_dummies = corrupcao_dummies.drop(columns=[corrupcao['regiao'].value_counts().idxmax()])
corrupcao_dummies.rename(columns={
      "América do Sul": 'America_do_sul',
      "EUA e Canadá" : 'EUA_Canada',
      "Ásia": "Asia"}, inplace = True)
corrupcao_dummies
```

```python
# Estimar o modelo de regressão múltipla com intercepto automático
modelo_paises = smf.ols(formula='cpi ~ America_do_sul+Oceania+EUA_Canada+Asia', data=corrupcao_dummies).fit()
modelo_paises.summary()
```
